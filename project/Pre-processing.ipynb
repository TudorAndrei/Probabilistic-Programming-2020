```python
import pandas as pd
import nltk
import gensim
from gensim.parsing.preprocessing import STOPWORDS
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np
np.random.seed(42)
from collections import Counter
from gensim.utils import simple_preprocess
from nltk.stem import WordNetLemmatizer, SnowballStemmer
```

```python
nltk.download('wordnet')
```

```python
my_stop_words = STOPWORDS.union(set(['say']))

def lemmatize_stemming(text):
    stemmer = SnowballStemmer(language='english')
    return stemmer.stem(WordNetLemmatizer().lemmatize(text, pos='v'))

def lemmatize(text):
    return WordNetLemmatizer().lemmatize(text, pos='v')

def preprocess(text, action="lemmatization_and_stemming"):
    result = []
    for token in gensim.utils.simple_preprocess(text):
        if token not in my_stop_words and len(token) > 3:
            if action == 'lemmatization_and_stemming':
                processed_token =lemmatize_stemming(token)
            else:
                processed_token = lemmatize(token)
            if processed_token not in my_stop_words:
                result.append(processed_token)
    return result
```
